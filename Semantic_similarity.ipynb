```
# semantic similarity

import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import LabelEncoder
np.random.seed(3)
```

```
df = pd.read_csv("SNLI_Corpus/snli_1.0_train.csv")
df = df[df['similarity']!= '-']
df = df.sample(n = 20000)
df = df.astype('str')
```

```
# converting to count tokenizer
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
cvr = TfidfVectorizer()
temp = cvr.fit_transform(pd.concat([df['sentence1'],df['sentence2']],axis = 0))
print(cvr.get_feature_names())
print('dict size:', len(cvr.get_feature_names()))
X1 = cvr.transform(df['sentence1']).toarray()
X2 = cvr.transform(df['sentence2']).toarray()
X = np.concatenate((X1,X2),axis = -1)

```

```
# target values 

lbl = LabelEncoder()
Y = lbl.fit_transform(df['similarity'])



```

```
#train and test split
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.05)


# fitting using logisitc regressio
# from sklearn.linear_model import LogisticRegression


# lr = LogisticRegression(verbose = True)
# lr.fit(x_train,y_train)
```

```
# keras tensorflow 

import tensorflow as tf
import keras
```

```
inp1 =   tf.keras.Input(shape= (X1.shape[1],))
inp2 =   tf.keras.Input(shape= (X2.shape[1],))
concat = tf.keras.layers.concatenate([inp1,inp2], axis=-1)
#im1 = tf.keras.layers.Dense(256,activation='relu')(concat)
output = tf.keras.layers.Dense(3,activation='softmax')(concat)

similar_sent =  keras.Model(inputs = [inp1,inp2],outputs = output)

```
